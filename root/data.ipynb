```python

ratios = []
results =[]

for x in range(0,9):
    df = pd.read_csv(f"Output/output_GoG_4a/{x}.txt")
    ratios.append(float(df[' b/c'].mean()))
    results.append(len(df[df[' result']==100])/1000)

print(ratios)
print(results)
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(ratios, results, c='blue', label = 'k=4')
ax.set_xlabel('b/c ratio')
ax.set_ylabel('Fixation probability')
ax.set_title('Death-birth')
ax.legend()
plt.show()
```
