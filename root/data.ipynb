```python
fig, ax = plt.subplots(figsize=(8,6))
markers = ['.', 's', '*', '+', 'X']
m = 0
for y in [3,4,6,8,10]:
    ratios=[]
    results=[]

    for x in range(0,9):
        df = pd.read_csv(f"Output/output_GoG_{y}/{x}.txt")
        ratios.append(float(df[' b/c'].mean()))
        results.append(len(df[df[' result']==100])/100000)

    label = 'k='+str(y)
    ax.scatter(ratios, results, label = label, marker = markers[m])
    m+=1

ax.scatter(3, 0.01, color='blue', marker=6, s=200)
ax.scatter(4, 0.01, color='orange', marker=6, s=200)
ax.scatter(6, 0.01, color='green', marker=6, s=200)
ax.scatter(8, 0.01, color='red', marker=6, s=200)
ax.scatter(10, 0.01, color='purple', marker=6, s=200)
# , c='blue'
ax.set_xlabel('Benefit-to-cost Ratio (b/c)')
ax.set_ylabel('Fixation probability')
# ax.set_title('Death-birth')
plt.axhline(y=0.01, color='r', linestyle='--')
ax.legend()
plt.savefig('../img/simulation-data.png')
plt.show()
```
